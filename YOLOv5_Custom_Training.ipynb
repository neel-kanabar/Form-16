```python
MY_KEY = "jNXm8Ra5p28TsjUZZRpl"
img_path = "/content/Form-16.jpeg"
```

```python
import requests
import base64
import io
from PIL import Image
```

```python
# Load Image with PIL
image = Image.open(img_path).convert("RGB")

# Convert to JPEG Buffer
buffered = io.BytesIO()
image.save(buffered, quality=90, format="JPEG")

# Base 64 Encode
img_str = base64.b64encode(buffered.getvalue())
img_str = img_str.decode("ascii")


# Construct the URL
upload_url = "".join([
    "https://detect.roboflow.com/form-16-gqtvl/3",
    "?api_key="+MY_KEY,
    "&name="+img_path
])
# POST to the API
r = requests.post(upload_url, data=img_str, headers={
    "Content-Type": "application/x-www-form-urlencoded"
})

# Output result
print(r.json())
```
